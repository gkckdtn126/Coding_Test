```python
bridge_length = 100
weight = 100
truck_weights = [10,10,10,10,10,10,10,10,10,10]
def solution(bridge_length, weight, truck_weights):
    answer = 0
    ing = []
    finish = []
    a = len(truck_weights)
    last_time = 0
    while True:
        past_time = answer
        while last_time-past_time != bridge_length:
            if len(truck_weights) != 0:
                    front = truck_weights.pop(0)
                    ing.append(front)
                    if sum(ing) <= weight:
                        print('real')
                        answer += 1
                    else:
                        truck_weights.insert(0,front)
                        ing.pop(-1)
                        answer += 1
            answer += 1
            print("ing",ing,answer)
            last_time = answer
            print(past_time)
            print(last_time)
        A = ing.pop(0)
        finish.append(A)
        print('finish',finish,answer)
        if len(finish) == a:
            return answer
```
